# DB 연결

```python
import pymysql

# MariaDB에 연결
mydb = pymysql.connect(
    host="localhost",
    user="root",
    password="1234",
    database="recipe"
)

mycursor = mydb.cursor()

# sql = "SELECT ingredients FROM new_content"

sql = "SELECT id,title,ingredients FROM new_table"
try:
    # 데이터베이스에 쿼리 실행
    mycursor.execute(sql)

    results = mycursor.fetchall()
    
except Exception as e:
    print("Error occurred:", e)

# 연결 종료
mydb.close()
```

```python
ingredients = []

for name in results:
    print(name)
    temp = name[2].strip('[]').split(', ')
    print(temp)
    for temp_name in temp:
        real_str = temp_name.strip('\'').split(' ')[:-1]
        temp_str = ''
        for element in real_str:
            temp_str += (element + ' ')
        if temp_str not in ingredients:
            ingredients.append(temp_str.strip())

# print(ingredients)
ingre= list(set(ingredients))
print(ingre)
```

# 단어 빈도수 확인

```python
from collections import Counter

result = Counter(ingredients)
result
```

```python
total_recipe = []

for name2 in results:
    temp2 = name2[2].strip('[]').split(', ')
    ingredients2 = []
    for temp_name2 in temp2:
        real_str2 = temp_name2.strip('\'').split(' ')[:-1]
        temp_str2 = ''
        for element in real_str2:
            temp_str2 += (element + ' ')
        if temp_str2 not in ingredients2:
            ingredients2.append(temp_str2.strip())

    name2_list = list(name2)
    name2_list.append(ingredients2)
    del name2_list[2]
    total_recipe.append(name2_list)

print(total_recipe)
```

# 재료별로 가중치 각각 부여

```python
final_dict = {}

for element in total_recipe:
    temp = []
    for value in element[2]:
        temp_dict = {}
        if value in element[1]:
            temp_dict[value] = 100
            temp.append(temp_dict)
        else:
            temp_dict[value] = 0.1
            temp.append(temp_dict)

    final_dict[element[1]] = temp
    
final_dict
```

# 메뉴별 가중치가 적용된 Dataframe CSV로 내보내기

```python
import pandas as pd

df = pd.DataFrame(columns=result)

for food, ingredients_list in final_dict.items():
    temp_df = pd.DataFrame(columns=df.columns)  # 빈 데이터프레임 생성
    for ingredient_dict in ingredients_list:
        for ingredient, count in ingredient_dict.items():
            temp_df.at[0, ingredient] = count  # 해당 재료의 카운트를 temp_df에 추가
    temp_df.index = [food]  # 행 인덱스를 음식 이름으로 설정
    df = pd.concat([df, temp_df])

df.fillna(0, inplace=True)
df.to_csv('NJG_BE/similarity.csv')
```
